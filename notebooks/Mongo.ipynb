```python
import pandas as pd
import pymongo 
```

```python
df_cinema_award = pd.read_csv("../data/Cinema Awards.csv", index_col=0)
```

```python
# Conect with Mongo

client = pymongo.MongoClient()
client
```

```python
# Check our databases

client.list_database_names()
```

```python
# Create a new database on Mongo DB

mydb = client["Cinema_Awards_db"]
```

```python
# Create a brand new collectionion in our database

collection = mydb.details
```

```python
# Insert data from our dataframe

collection.insert_many(df_cinema_award.to_dict("records"))
```

```python
# Pass the query parameter without any specification, so it will return all the elements of the database

collection.find()
```

```python
list(collection.find())[3]
```

```python
collection.distinct( "film")
```

```python
# Find the most oscar awarded movies

most_oscar = pd.DataFrame(collection.find({'oscar': {"$eq": "Winner"}}))
most_oscar.name.unique()
```
